```python
# 1. Install Hugging Face Transformers
pip install transformers

# 2. Import necessary libraries
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# 3. Load the pre-trained FinBERT model and tokenizer
# Specifically the one from Huang, Wang, and Yang (yya518)
model_name = "yiyanghkust/finbert-tone" # or 'ProsusAI/finbert'
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=3)
tokenizer = BertTokenizer.from_pretrained(model_name)

# 4. Define your financial text
sentence = "there is a shortage of capital and we need extra financing."

# 5. Tokenize and predict
inputs = tokenizer(sentence, return_tensors="pt")
outputs = model(**inputs)
predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)

# 6. Map results
labels = ['positive', 'negative', 'neutral']
print(labels[torch.argmax(predictions)])
```

### Trying an alternative FinBERT model: `ProsusAI/finbert`

```python
model_name_alt = "ProsusAI/finbert"
model_alt = BertForSequenceClassification.from_pretrained(model_name_alt, num_labels=3)
tokenizer_alt = BertTokenizer.from_pretrained(model_name_alt)

inputs_alt = tokenizer_alt(sentence, return_tensors="pt")
outputs_alt = model_alt(**inputs_alt)
predictions_alt = torch.nn.functional.softmax(outputs_alt.logits, dim=-1)

labels_alt = ['positive', 'negative', 'neutral']
predicted_sentiment_alt = labels_alt[torch.argmax(predictions_alt)]

print(f"Original sentence: '{sentence}'")
print(f"Sentiment from {model_name}: {labels[torch.argmax(predictions)]}")
print(f"Sentiment from {model_name_alt}: {predicted_sentiment_alt}")
```
